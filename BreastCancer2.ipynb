```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_refined.csv')
df.drop('Unnamed: 0', axis = 1, inplace = True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['diagnosis']= encoder.fit_transform(df['diagnosis'])
df
```

```python
bc_df = df[['diagnosis', 'radius_mean', 'perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean', 'radius_worst',
           'perimeter_worst', 'area_worst', 'concavity_worst', 'concave points_worst']]
bc_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_columns = 10
X = bc_df.iloc[:, 1:x_columns].values
y = bc_df['diagnosis'].values
y
train_ratio = 0.80
test_ratio = 0.20
validation_ratio = 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_ratio)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = validation_ratio)
print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
K = 5
k_model = KNeighborsClassifier(n_neighbors=K)
k_model.fit(X_train, y_train)
y_pred = k_model.predict(X_test)
y_pred
k_score = k_model.score(X_test, y_test)
k_score
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel = 'linear', max_iter = 100)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
y_pred
svm_score = svm_model.score(X_test, y_test)
svm_score
```
